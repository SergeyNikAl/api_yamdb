```python
# from reviews.models import Category, Comment, Genre, Review, Title, User
import csv


MAPPING = {'author': 'author_id'}

csv_folder_path = 'data/'
csvs_models = {
    'category.csv': Category,
    'comments.csv': Comment,
    'genre.csv': Genre,
    'review.csv': Review,
    'titles.csv': Title,
    'users.csv': User
}

csv_files = ['review.csv']

genre_title = 'genre_title.csv'


def save_model(row_dict, model):
    to_save = model(**row_dict)
    to_save.save()

def load_csv():
    for file_name, model in csv_models.items():
        with open(csv_folder_path + file_name) as file:
            reader = csv.reader(file)
            headline = next(reader)
            headline = [MAPPING.get(col, col) for col in headline]
            for row in reader:
                row_dict = {name:value for name, value in zip(headline, row)}
                save_model(row_dict, model)
        break
```
